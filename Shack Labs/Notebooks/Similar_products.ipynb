## Preprocessing Datasets:

- Here I'm removing some columns that are not going to help in retrieving information.
- Rather than cleaning tons of rows; only going to select columns which are important, as mentioned in the pdf.

- Further I'm converting uppercase/capitalized words/strings into lowercase format.

```python
def delete_redundant_columns(column_names: list, data: pd.DataFrame) -> None:
    
    """ removes redundant columns and returns the dataset"""
    
    if len(column_names) == 0:
        print(f"enter column names in the list to remove from the dataset.")
        return
    data.drop(columns=column_names, inplace=True)
    
    return
```

```python
def get_lowercase_text(column: str, data: pd.DataFrame) -> None:
    
    """ converts capital/uppercase strings or text into lowercase. """
    
    if column is None:
        print(f"empty column name")
        return 
    if data.shape[0] == 0:
        print(f"empty dataset: {len(data)}")
    
    data[column] = data[column].apply(lambda x: " ".join(x.lower() for x in x.split()))
    
    return
```

## Hypothesis:

- As mentioned in the problem statement that these two datasets has their products in common.
- What I want to do here is to make sure that; there are any similar products in given datasets.



- So my hypothesis here is that (also mentioned in the problem statement) there are similar products in given two datasets.
- If not we reject our hypothesis.


## Proof:

- To make sure that the given datasets has similar products, I'll use set data datastructure to check my hypothesis, I will save both dataset's product names saparately and use intersection method on those two sets to get all the common products names.
- Reason behind using set is because rather than iterating through all the products and comparing one by one can lead to great computation and time complexity, instead set methods are efficient and great alternative in case of our problem.


- Since there's a lot of repetition of same product, we are going to get only one occurance of that product from both the datasets, as set intersection method gives us only the common elements from two collections.

```python
amazon_prod_set = set(amazon.amz_product.to_list())
flipkart_prod_set = set(flipkart.product_name.to_list())

# -- Gives all the common products from both sets -- 
# list(amazon_prod_set.intersection(flipkart_prod_set))   

len(amazon_prod_set.intersection(flipkart_prod_set))
```

### Note:

- Here we can see that there are total 12604 unique products that are common in both datasets, and these products are repeating themselves again and again yet with different retatil and discounted prices.
- We've proved aur hypothesis.


- Below I'm quickly going to check some products which can be available in above two sets.

```python
def check_product(product_name: str, products_set: set) -> bool:
    
    """ returns a boolean value if product is present or not """
    
    if product_name == "":
        print("enter valid product name.")
        return
    if len(products_set) == 0:
        print("empty set given.")
        return
    
    if product_name in products_set:
        return True
    
    return False
```

```python
product_names = ["alisha solid women's cycling shorts", 
                 "alisha solid women's cycling ", 
                 "faith leenise 1000137 women heels"
                ]

for prod in product_names:
    print(f"{prod} : amazon products set : {check_product(prod, amazon_prod_set)}\n"
          f"{prod} : flipkart products set : {check_product(prod, amazon_prod_set)}\n"
         )
```

## Combining amazon & flipkart dataset for comparison:
